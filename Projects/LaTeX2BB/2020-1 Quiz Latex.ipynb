```python
############################################description############################################
# 모든 문제가 ox 일 때 사용 가능한 코드입니다.
# 수식에서 모든 띄어쓰기를 제거(partial 예외)했기 때문에 수식은 검토가 필요합니다.
# 수식 띄어쓰기 검토만 하고 바로 블랙보드 업로드가 가능합니다.
#
# 이 코드는 다음과 같은 가정을 가지고 만들어졌습니다.
# 1. \question 으로 전체 문제 설명, 이 문제에 대해 여러 새끼문제는 모두  \begin{part} ~ \end{part} 의 꼴을 띈다.
# 2. 단일문제의 경우 \question 으로 표현. 
# 3. LaTeX 문법에 오류가 전혀 없을 것. => A[i]=='$' 에서 오류가 생기는 경우, 부분별로 넣어가며 LaTeX 코드의 오류를 찾아야 합니다.
#
############################################description############################################

B=str(input())
B=B.replace('\\','Z')
############################################Z로 바꿔야 전환이 먹힘.
B=B.replace('Z[','$')
B=B.replace('Z]','$')
B=B.replace('Zbegin{equation}','$')
B=B.replace('Zend{equation}','$')
B=B.replace('Zbegin{eqnarray*}','$')
B=B.replace('Zend{eqnarray*}','$')
B=B.replace('Zbegin{enumerate}','Zbegin{parts}')
B=B.replace('Zend{enumerate}','Zend{parts}')
B=B.replace('Zitem','Zpart')
B=B.replace('  ',' ')
############################################
B=B.replace('Z','\\')
B=B.replace('\t','')
############################################문장에서 \\ 제거, 두번 띄어쓰기 제거, 수식에서 띄어쓰기 모두 삭제
A=list(B)
K=False
i=0
while i<len(A)-1:
    if A[i]=='$':
        K=not K
    elif A[i]=='\\' and A[i+1]=='\\' and K==False:
        del A[i]
        del A[i]
        i=i-1
    elif A[i]==' ' and A[i+1]==' ' and K==False:
        del A[i]
        i=i-1
    while K==True:
        i=i+1
        if A[i]=='$':
            K=not K
        elif A[i]==' ':
            del A[i]
            i=i-1
        elif A[i]=='\t':
            del A[i]
            i=i-1
    i=i+1
i=0
while i<len(A)-1:
    if A[i]==' ' and A[i+1]==' ':
        del A[i]
        i=i-1
    i=i+1
########################################################################################문제 쪼개기
B = ''.join(A)
B=B.replace('\\','Z')
B=B.replace('Zpartial',' QWERTY ')
index=[0]
for i in range(B.count('Zquestion')):
    index.append(B.find('Zquestion',index[i]))
    B=B[:index[i+1]]+B[index[i+1]+1:]
del index[0]
index.append(len(B))
def gtn(x):
    Temp2=x[0:x.find('begin{parts}')-1]
    Temp3=x[x.find('begin{parts}')+12:]
    Temp=Temp3.replace('Zpart','TF '+Temp2)
    return Temp

def ftn(K,x):
    if x.find('begin{parts}')>0:
        return K+gtn(x)
    else:
        return K+' TF '+x
        
K=''
for i in range(len(index)-1):
    Temp=''
    for j in range(index[i]+9,index[i+1]-1):
        Temp=Temp+B[j]
    K=ftn(K,Temp)
    
K=K.replace(' QWERTY ','Zpartial ')  
########################################################################################begin{center} ~ end{center} 제거
for i in range(K.count('begin{center}')):
    ind1=K.find('Zbegin{center}')
    ind2=K.find('Zend{center}')
    K=K[:ind1]+K[ind2+12:]
    
########################################################################################$Zbigcirc$ 뒤에 설명 제거
ind1=[0]; ind2=[]
for i in range(K.count('$Zbigcirc$')):
    ind1.append(K.find('$Zbigcirc$',ind1[i]+1))
    ind2.append(K.find('TF',ind1[i+1]))
del ind1[0]
if -1 in ind2:
    ind2.remove(-1)
ind1.append(len(K))
ind2.append(len(K))
ind1.reverse()
ind2.reverse()
for i in range(len(ind1)):
    K=K[:ind1[i]+11]+K[ind2[i]:]    

########################################################################################$Ztimes$ 뒤에 설명 제거
ind1=[0]; ind2=[]
for i in range(K.count('$Ztimes$')):
    ind1.append(K.find('$Ztimes$',ind1[i]+1))
    ind2.append(K.find('TF',ind1[i+1]))
del ind1[0]
if -1 in ind2:
    ind2.remove(-1)
ind1.append(len(K))
ind2.append(len(K))
ind1.reverse()
ind2.reverse()
for i in range(len(ind1)):
    K=K[:ind1[i]+9]+K[ind2[i]:]    

K=K.replace('Z','\\')
########################################################################################간격 다시 제거
A=list(K)
K=False
i=0
while i<len(A)-1:
    if A[i]=='$':
        K=not K
    elif A[i]=='\\' and A[i+1]=='\\' and K==False:
        del A[i]
        del A[i]
        i=i-1
    elif A[i]==' ' and A[i+1]==' ' and K==False:
        del A[i]
        i=i-1
    i=i+1
i=0
while i<len(A)-1:
    if A[i]==' ' and A[i+1]==' ':
        del A[i]
        i=i-1
    i=i+1
########################################################################################
st=''.join(A)
st=st.replace('$','$$')
st=st.replace('&=&','=')
############################################
st=st.replace('\\','Z')
############################################
st=st.replace('Zbegin{parts}','')
st=st.replace('Zend{parts}','')
st=st.replace('Zend{parts','')

st=st.replace('Zmathbbm','Zmathbb')
st=st.replace('Zquestion','TF')
st=st.replace('TF','\nTF\t')
st=st.replace('TF ','\nTF\t')

st=st.replace('Ztextbf{A:}','')
st=st.replace('$$Ztimes$$','\t'+'false')
st=st.replace('$$Zbigcirc$$','\t'+'true')
st=st.replace('Z','\\')
########################################################################################
print('='*20,'Starts From Here','='*20)
print(st)
```

```python
########################################################################################
# Chapter 1 만을 위한 긴급조치로, 다른 단원에 사용될 수 없습니다.
# 수정일자 20.03.23
########################################################################################
B=str(input())
B=B.replace('\\','Z')
############################################Z로 바꿔야 전환이 먹힘.
B=B.replace('Z[','$')
B=B.replace('Z]','$')
B=B.replace('Zbegin{equation}','$')
B=B.replace('Zend{equation}','$')
B=B.replace('Zbegin{eqnarray*}','$')
B=B.replace('Zend{eqnarray*}','$')
B=B.replace('Zquestion[1]','Zquestion ')
B=B.replace('Zquestion[2]','Zquestion ')
B=B.replace('Zquestion[3]','Zquestion ')
B=B.replace('Zquestion[4]','Zquestion ')
B=B.replace('Zquestion[5]','Zquestion ')
B=B.replace('Zquestion[6]','Zquestion ')
B=B.replace('Zquestion[7]','Zquestion ')
B=B.replace('Zquestion[8]','Zquestion ')
B=B.replace('Zquestion[9]','Zquestion ')
B=B.replace('Zquestion[10]','Zquestion ')
B=B.replace('Zquestion[11]','Zquestion ')
B=B.replace('Zitem','Zquestion ')
B=B.replace('  ',' ')
############################################
B=B.replace('Z','\\')
B=B.replace('\t','')
############################################문장에서 \\ 제거, 두번 띄어쓰기 제거, 수식에서 띄어쓰기 모두 삭제
A=list(B)
K=False
i=0
while i<len(A)-1:
    if A[i]=='$':
        K=not K
    elif A[i]=='\\' and A[i+1]=='\\' and K==False:
        del A[i]
        del A[i]
        i=i-1
    elif A[i]==' ' and A[i+1]==' ' and K==False:
        del A[i]
        i=i-1
    while K==True:
        i=i+1
        if A[i]=='$':
            K=not K
        elif A[i]==' ' and A[i+1]==' ':
            del A[i]
            i=i-1
        elif A[i]=='\t':
            del A[i]
            i=i-1
    i=i+1
i=0
while i<len(A)-1:
    if A[i]==' ' and A[i+1]==' ':
        del A[i]
        i=i-1
    i=i+1
########################################################################################문제 쪼개기
B = ''.join(A)
B=B.replace('\\','Z')
B=B.replace('Zpartial',' QWERTY ')
index=[0]
for i in range(B.count('Zquestion')):
    index.append(B.find('Zquestion',index[i]))
    B=B[:index[i+1]]+B[index[i+1]+1:]
del index[0]
index.append(len(B))
def gtn(x):
    Temp2=x[0:x.find('begin{parts}')-1]
    Temp3=x[x.find('begin{parts}')+12:]
    Temp=Temp3.replace('Zpart','TFQN '+Temp2)
    return Temp

def ftn(K,x):
    if x.find('begin{parts}')>0:
        return K+gtn(x)
    else:
        return K+' TFQN '+x
        
K=''
for i in range(len(index)-1):
    Temp=''
    for j in range(index[i]+9,index[i+1]-1):
        Temp=Temp+B[j]
    K=ftn(K,Temp)
    
K=K.replace(' QWERTY ','Zpartial ') 
########################################################################################begin{center} ~ end{center} 제거
for i in range(K.count('begin{center}')):
    ind1=K.find('Zbegin{center}')
    ind2=K.find('Zend{center}')
    K=K[:ind1]+K[ind2+12:]
    
########################################################################################$Ztimes$ 뒤에 설명 제거
ind1=[0]; ind2=[]
for i in range(K.count('$Ztimes$')):
    ind1.append(K.find('$Ztimes$',ind1[i]+1))
    ind2.append(K.find('TFQN',ind1[i+1]))
del ind1[0]
if -1 in ind2:
    ind2.remove(-1)
ind1.append(len(K))
ind2.append(len(K))
ind1.reverse()
ind2.reverse()
for i in range(len(ind1)):
    K=K[:ind1[i]+9]+K[ind2[i]:]    

K=K.replace('Z','\\')
K=K.replace('.$','$.')
K=K.replace('.','. ')
########################################################################################간격 다시 제거
A=list(K)
K=False
i=0
while i<len(A)-1:
    if A[i]=='$':
        K=not K
    elif A[i]=='\\' and A[i+1]=='\\' and K==False:
        del A[i]
        del A[i]
        i=i-1
    elif A[i]==' ' and A[i+1]==' ' and K==False:
        del A[i]
        i=i-1
    i=i+1
i=0
while i<len(A)-1:
    if A[i]==' ' and A[i+1]==' ':
        del A[i]
        i=i-1
    i=i+1
########################################################################################
st=''.join(A)
st=st.replace('$','$$')
st=st.replace('&=&','=')
############################################
st=st.replace('\\','Z')
############################################
st=st.replace('Zbegin{parts}','')
st=st.replace('Zend{parts}','')
st=st.replace('Zend{parts','')

st=st.replace('Zmathbbm','Zmathbb')
st=st.replace('Zquestion','TF')
st=st.replace('TFQN','\nMC\t')
st=st.replace('TFQN ','\nMC\t')

st=st.replace('Ztextbf{A:}','')
st=st.replace('$$Ztimes$$','\t'+'true'+'\t'+'incorrect'+'\t'+'false'+'\t'+'correct'+'\t'+'pass'+'\t'+'incorrect')
st=st.replace('$$Zbigcirc$$','\t'+'true'+'\t'+'correct'+'\t'+'false'+'\t'+'incorrect'+'\t'+'pass'+'\t'+'incorrect')
st=st.replace('Z','\\')
########################################################################################
print('='*20,'Starts From Here','='*20)
print(st)
```
