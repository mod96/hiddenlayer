```python
##############################################Description##############################################
# 본 코드는 블랙보드에 올라간 LaTeX 코드를 가동시키고 부분점수를 부여합니다.
# 사용시 변수들을 채워넣고 실행시키면 됩니다.
# selenium, bs4 는 pip install 보다는 conda install 하는 것이 좋습니다. (collab 에서는 그냥 될수도? - 확인 안해봄)
# chrome driver 가 필요합니다. ghostdriver 도 좋지만 selenium 에서 곧 삭제한다 본 것 같음
# headless option 사용시 LaTeX 코드가 제대로 가동되지 않습니다. 적절한 option을 추가하면 될 법 하지만 귀찮습니다.ㅎㅎ
# time, winsound 는 기본 패키지 입니다. 가동이 끝나면 삐- 소리가 날겁니다.
# 이 코드로 한번 가동시킨 문제는 다시 시행시 오류를 내보냅니다. 분석하기 귀찮아서 안함. 그냥 시작점 새로 정하고 하세염
##############################################Description##############################################
###############################################################변수들
MAIN_ID=' ' #BB 아이디
MAIN_PW=' ' #BB 비번
MAIN_URL='https://kulms.korea.ac.kr/webapps/assessment/do/content/assessment?action=MODIFY&course_id=_186329_1&content_id=_3160638_1&assessmentType=Test&method=modifyAssessment'
#'과제 -> 원하는 문제의 "수정"' 의 페이지 주소
DRIVER_LOCATION='C:/Users/mod96/Desktop/HSH/Programs/WebCrawling/chromedriver.exe'
StartingProbNUM=88
EndingProbNUM=100
###############################################################아래부터 건들지 말것
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import winsound as ws
Log_IN = 'https://kulms.korea.ac.kr'
driver=webdriver.Chrome(DRIVER_LOCATION)
driver.get(Log_IN)
time.sleep(1)
entt=webdriver.common.keys.Keys.ENTER
ctrll=webdriver.common.keys.Keys.CONTROL
erase=webdriver.common.keys.Keys.BACKSPACE
############################################################### 로그인부터 시험수정까지 들어가는 action

driver.find_element_by_xpath('//*[@id="one_id"]').send_keys(MAIN_ID)
driver.find_element_by_xpath('//*[@type="password"]').send_keys(MAIN_PW)
driver.find_element_by_xpath('//*[@type="password"]').send_keys(entt)

time.sleep(1)
driver.get(MAIN_URL) #kulms 에서 로그인 하고 바로 시험 수정 페이지로 넘어가는 겁니다.
time.sleep(1)
############################################################### 문제별 수정작업 시작 - 여기서 문제가 날수도 있는데 보고 바람
try:
    for i in range(int(StartingProbNUM)-1,int(EndingProbNUM)):
        num=str(i+1)
        driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/div/div/div/div[3]/form/ul/li['+num+']/div[1]/span/a').send_
        keys(entt)
        driver.find_element_by_xpath('//*[@title="수정"]').send_keys(entt)

        btn=driver.find_element_by_xpath('//*[@id="partialCreditId"]')
        driver.execute_script("arguments[0].click();", btn)

        btn=driver.find_element_by_xpath('//*[@id="allowNegativeScoresId"]')
        driver.execute_script("arguments[0].click();", btn)

        html = driver.page_source 
        soup = BeautifulSoup(html, 'html.parser') 
        notices = soup.select('tbody.answerBlock')
        for item in notices:
            k=str(item)
            if k.find('checked')>0 and k.find('정답 1')>0:
                bull=False
                break
            elif k.find('checked')>0 and k.find('정답 2')>0:
                bull=True
                break

        if bull:
            driver.find_element_by_xpath('//*[@id="partialCreditPoints[0]"]').send_keys(ctrll,'a')
            driver.find_element_by_xpath('//*[@id="partialCreditPoints[0]"]').send_keys(erase)
            driver.find_element_by_xpath('//*[@id="partialCreditPoints[0]"]').send_keys(-50)
        else:
            driver.find_element_by_xpath('//*[@id="partialCreditPoints[1]"]').send_keys(ctrll,'a')
            driver.find_element_by_xpath('//*[@id="partialCreditPoints[1]"]').send_keys(erase)
            driver.find_element_by_xpath('//*[@id="partialCreditPoints[1]"]').send_keys(-50)
        time.sleep(1)
        btn=driver.find_element_by_xpath('//*[@class="submit button-1"]')
        driver.execute_script("arguments[0].click();", btn)
except:
    print(ws.Beep(2000, 5000))
    print("error at:",i)
    

print(ws.Beep(2000, 1000))
```
